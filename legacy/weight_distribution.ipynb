```python
import math
# generate binomial coefficients
def binomial_coefficients(n):
    """Generate binomial coefficients C(n, k) for k = 0, 1, ..., n."""
    return [np.math.comb(n, k) for k in range(1, n + 1)]

def upper_bound_binomial_coefficients(n):
    # return [math.factorial(n)/(math.factorial(k)*math.factorial(n-k)) for k in range(1, n + 1)]
    return [(math.e*n/k)**k for k in range(1, n + 1)]

# plot binomial coefficients with n=10 for k=0,1,...,10
fig, ax = plt.subplots(figsize=(6, 4))
n = 10
k = np.arange(1, n + 1)
coeffs = binomial_coefficients(n)
bound_coeffs = upper_bound_binomial_coefficients(n)
ax.plot(k[:5], coeffs[:5], marker='o', linestyle='-', color=cc[0])
ax.plot(k[:5], bound_coeffs[:5], marker='.', linestyle='-', color=cc[0])
ax.set_title(f'Binomial Coefficients C({n}, k)')
ax.set_xlabel('k')  
ax.set_ylabel('C(n, k)')
```
